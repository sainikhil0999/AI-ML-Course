```python

# --------------------------------------------------
# 2. using constructor
# --------------------------------------------------

class Employee:
    def __init__(self, name, department, year_of_joining):
        self.name = name
        self.department = department
        self.year_of_joining = year_of_joining

emp2 = Employee("Kiran", "HR", 2020)

print("2. constructor example")
print(emp2)
print("%s works in %s and joined in %d." %
      (emp2.name, emp2.department, emp2.year_of_joining))


```

```python

# --------------------------------------------------
# 3. methods in class
# --------------------------------------------------

class Employee:
    def __init__(self, name, department, year_of_joining):
        self.name = name
        self.department = department
        self.year_of_joining = year_of_joining

    def years_in_company(self, current_year):
        return current_year - self.year_of_joining

    def next_year_service(self, current_year):
        return self.years_in_company(current_year) + 1

    def __str__(self):
        return "%s works in %s and joined in %d." % (
            self.name, self.department, self.year_of_joining
        )

emp3 = Employee("Meena", "Finance", 2019)
emp4 = Employee("Arun", "Support", 2022)

print("3. methods example")
print(emp3)
print("service years:", emp3.years_in_company(2026))
print("next year service:", emp3.next_year_service(2026))
print()
print(emp4)
print("service years:", emp4.years_in_company(2026))
print("next year service:", emp4.next_year_service(2026))


```

```python

# --------------------------------------------------
# 4. class without __str__
# --------------------------------------------------

class Product:
    def __init__(self, item_name, price):
        self.item_name = item_name
        self.price = price

    def discount_price(self):
        return self.price - 100

p1 = Product("Headphones", 2500)

print("4. without __str__")
print(p1)
print("discount price:", p1.discount_price())
```

```python

# --------------------------------------------------
# 5. setter style methods
# --------------------------------------------------

class Customer:
    def set_name(self, name):
        self.name = name

    def set_city(self, city):
        self.city = city

    def set_age(self, age):
        self.age = age

    def show_data(self):
        return "%s lives in %s and age is %d." % (
            self.name, self.city, self.age
        )

c1 = Customer()
c1.set_name("Nikhil")
c1.set_city("Dallas")
c1.set_age(27)

print("5. setter methods")
print(c1.show_data())
```

```python

# --------------------------------------------------
# 6. __str__ and __repr__
# --------------------------------------------------

class Movie:
    def __init__(self, title, hero, release_year):
        self.title = title
        self.hero = hero
        self.release_year = release_year

    def __str__(self):
        return f"{self.title} starring {self.hero} released in {self.release_year}."

    def __repr__(self):
        return f"Movie(title='{self.title}', hero='{self.hero}', release_year={self.release_year})"

m1 = Movie("Leo", "Vijay", 2023)

print("6. __str__ and __repr__")
print(m1.hero)
print("using str():", str(m1))
print("using repr():", repr(m1))


```

```python

# --------------------------------------------------
# 7. single underscore
# --------------------------------------------------

class BankAccount:
    def __init__(self, acc_holder, balance):
        self._acc_holder = acc_holder
        self._balance = balance

    def check_balance(self):
        return self._balance

    def __str__(self):
        return "%s has balance %d." % (self._acc_holder, self._balance)

b1 = BankAccount("Rahul", 45000)

print("7. single underscore")
print(b1)
print(b1._acc_holder)
print(b1._balance)
```

```python

# --------------------------------------------------
# 8. double underscore
# --------------------------------------------------

class BankAccount:
    def __init__(self, acc_holder, balance):
        self.__acc_holder = acc_holder
        self.__balance = balance

    def check_balance(self):
        return self.__balance

    def __str__(self):
        return "%s has balance %d." % (self.__acc_holder, self.__balance)

b2 = BankAccount("Sneha", 60000)

print("8. double underscore")
print(b2)
print(b2._BankAccount__acc_holder)
print(b2._BankAccount__balance)
print(b2.check_balance())
```

```python

# --------------------------------------------------
# 9. getters
# --------------------------------------------------

class BankAccount:
    def __init__(self, acc_holder, balance, branch):
        self.__acc_holder = acc_holder
        self.__balance = balance
        self.__branch = branch

    def get_acc_holder(self):
        return self.__acc_holder

    def get_balance(self):
        return self.__balance

    def get_branch(self):
        return self.__branch

    def __str__(self):
        return "%s belongs to %s branch and has %d balance." % (
            self.__acc_holder, self.__branch, self.__balance
        )

b3 = BankAccount("Pooja", 75000, "Plano")

print("9. getters")
print(b3.get_acc_holder())
print(b3.get_balance())
print(b3.get_branch())
print(b3)
```

```python
# --------------------------------------------------
# 10. __dict__
# --------------------------------------------------

class Laptop:
    def __init__(self, brand, model, price):
        self._brand = brand
        self._model = model
        self._price = price

    def __str__(self):
        return "%s %s costs %d." % (self._brand, self._model, self._price)

lap1 = Laptop("Dell", "Inspiron", 58000)

print("10. __dict__")
print(lap1.__dict__)
print(lap1._brand)
print(lap1._model)
print(lap1)

lap1.__stock = 25
print(lap1.__dict__)


print("\n---------------------------------\n")
```

```python

# --------------------------------------------------
# 11. inheritance
# --------------------------------------------------

class Person:
    def __init__(self, first_name, last_name, birth_year):
        self._first_name = first_name
        self._last_name = last_name
        self._birth_year = birth_year

    def full_name(self):
        return self._first_name + " " + self._last_name

    def age(self, current_year):
        return current_year - self._birth_year


class Student(Person):
    def __init__(self, student_id, course, *args):
        super(Student, self).__init__(*args)
        self._student_id = student_id
        self._course = course

    def show_student(self):
        print("student id:", self._student_id)
        print("name:", self.full_name())
        print("course:", self._course)

s1 = Student(101, "Python", "Sai", "Kumar", 2002)

print("11. inheritance")
print(s1._student_id)
print(type(s1))
print(s1._first_name)
print(s1._last_name)
print(isinstance(s1, Person))
print(isinstance(s1, object))
s1.show_student()
print("age:", s1.age(2026))
```

```python
# --------------------------------------------------
# 12. one more inheritance example
# --------------------------------------------------

class Vehicle:
    def __init__(self, brand, year):
        self.brand = brand
        self.year = year

    def vehicle_info(self):
        return "%s vehicle from %d" % (self.brand, self.year)


class Car(Vehicle):
    def __init__(self, seats, *args):
        super(Car, self).__init__(*args)
        self.seats = seats

    def show_car(self):
        print(self.vehicle_info())
        print("seats:", self.seats)

car1 = Car(5, "Toyota", 2021)

print("12. another inheritance example")
car1.show_car()
```

```python

# --------------------------------------------------
# 13. class variable and class method
# --------------------------------------------------

class College:
    college_name = "Global College"

    def __init__(self, student_name):
        self.student_name = student_name

    @classmethod
    def change_college_name(cls, new_name):
        cls.college_name = new_name

    def show_info(self):
        print("student:", self.student_name)
        print("college:", College.college_name)

st1 = College("Rakesh")
st2 = College("Divya")

print("13. class variable and class method")
st1.show_info()
print()
st2.show_info()

College.change_college_name("City College")

print("\nafter changing college name")
st1.show_info()
print()
st2.show_info()
```

```python
# --------------------------------------------------
# 14. dictionary inside class
# --------------------------------------------------

class Store:
    def __init__(self):
        self.items = {
            "milk": 60,
            "bread": 40,
            "eggs": 90
        }

    def show_items(self):
        print(self.items)

    def get_price(self, item):
        if item in self.items:
            print(item, "price is", self.items[item])
        else:
            print(item, "not available")

shop = Store()

print("14. dictionary inside class")
shop.show_items()
shop.get_price("bread")
shop.get_price("juice")

```

```python

# --------------------------------------------------
# 15. abstraction
# --------------------------------------------------

from abc import ABC, abstractmethod

class Payment(ABC):
    @abstractmethod
    def pay(self):
        pass

class UPI(Payment):
    def pay(self):
        print("payment done using upi")

class Card(Payment):
    def pay(self):
        print("payment done using card")

pay1 = UPI()
pay2 = Card()

print("15. abstraction")
pay1.pay()
pay2.pay()
```

```python
# --------------------------------------------------
# 16. polymorphism using *args
# --------------------------------------------------

class Calculator:
    def add(self, *args):
        total = 0
        for i in args:
            total += i
        print("sum is", total)

calc = Calculator()

print("16. polymorphism using *args")
calc.add(10, 20)
calc.add(10, 20, 30)
calc.add(1, 2, 3, 4, 5)

```

```python

# --------------------------------------------------
# 17. polymorphism using **kwargs
# --------------------------------------------------

class StudentInfo:
    def display(self, **kwargs):
        for key, value in kwargs.items():
            print(key, ":", value)

st = StudentInfo()

print("17. polymorphism using **kwargs")
st.display(name="Kiran", age=21)
print()
st.display(name="Lavanya", age=22, city="Hyderabad", course="Python")

```

```python

# --------------------------------------------------
# 18. using both *args and **kwargs
# --------------------------------------------------

class Demo:
    def show(self, *args, **kwargs):
        print("args values:", args)
        print("kwargs values:", kwargs)

d = Demo()

print("18. using both *args and **kwargs")
d.show(100, 200, 300, name="Sai", city="Dallas")


print("assignment completed")
```
