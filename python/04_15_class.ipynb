```python

# 1. functions

def add(a, b):                  # this function adds 2 numbers
    return a + b                # giving back the answer

x = add(10, 5)                  # calling the function
print("sum is", x)              # printing result

def greet(name):                # function to greet a person
    return "hello " + name      # joining hello with name

print(greet("nikhil"))          # function call# class notes python basics

```

```python
# 2. def statement

def student_details(name, age): # def is used to create a function
    print("name is", name)      # printing student name
    print("age is", age)        # printing student age

student_details("ravi", 22)     # calling the function

```

```python
# 3. return vs print

def using_print(a, b):          # this function uses print
    print(a + b)                # it only shows output

def using_return(a, b):         # this function uses return
    return a + b                # it sends value back

using_print(2, 3)               # will just print 5

ans = using_return(2, 3)        # storing returned value
print(ans)                      # printing stored value

print(using_return(4, 5) * 2)   # return can be used again in another operation
```

```python

# 4. iterators

nums = [10, 20, 30, 40]         # normal list
it = iter(nums)                 # converting list into iterator
5tyh
print(next(it))                 # first value
print(next(it))                 # second value
print(next(it))                 # third value
print(next(it))                 # fourth value

word = "python"                 # string example
it2 = iter(word)                # making iterator from string

print(next(it2))                # gives p
print(next(it2))                # gives y
```

```python
# 5. generators

def count_numbers(n):           # generator function
    for i in range(1, n + 1):   # loop from 1 to n
        yield i                 # giving values one by one

for value in count_numbers(5):  # calling generator
    print(value)                # printing each value

def give_even(n):               # generator for even numbers
    for i in range(2, n + 1, 2):
        yield i

for num in give_even(10):       # calling second generator
    print(num)                  # printing even numbers
```

```python
# 6. map

marks = [40, 50, 60, 70]        # list of marks
new_marks = list(map(lambda x: x + 5, marks))   # adding 5 to each mark

print(marks)                    # old marks
print(new_marks)                # updated marks

names = ["sai", "nikhil", "arun"]   # list of names
upper_names = list(map(str.upper, names))   # converting to uppercase

print(names)                    # original names
print(upper_names)              # capital names
```

```python
# 7. reduce

from functools import reduce    # importing reduce

values = [1, 2, 3, 4, 5]        # list of values
total = reduce(lambda a, b: a + b, values)   # adding all values

print(values)                   # original list
print("total is", total)        # final sum

nums2 = [2, 3, 4]               # another list
product = reduce(lambda a, b: a * b, nums2)  # multiplying values

print(nums2)                    # original list
print("product is", product)    # final product
```

```python
# 8. filter

numbers = [1, 2, 3, 4, 5, 6, 7, 8]   # list of numbers
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))   # keeping only even numbers

print(numbers)                  # full list
print(even_numbers)             # filtered even numbers

ages = [12, 18, 21, 15, 30]     # age list
adults = list(filter(lambda x: x >= 18, ages))   # keeping age 18 and above

print(ages)                     # original age list
print(adults)                   # filtered list

```

```python
# 9. file operations in python

file = open("notes.txt", "w")   # opening file in write mode
file.write("this is first line\n")   # writing line 1
file.write("this is second line\n")  # writing line 2
file.close()                    # closing file

file = open("notes.txt", "r")   # opening file in read mode
data = file.read()              # reading all content
file.close()                    # closing file
print(data)                     # printing file content

file = open("notes.txt", "a")   # opening in append mode
file.write("this is added later\n")  # adding one more line
file.close()                    # closing file

file = open("notes.txt", "r")   # opening again to check updated content
data2 = file.read()             # reading again
file.close()                    # closing file
print(data2)                    # printing updated content
```
