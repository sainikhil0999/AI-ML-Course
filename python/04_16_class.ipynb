```python
# topics done in class:
# modules(module means a python file with functions in it)
# packages(package means a folder with modules)
# libraries(library means collection of many modules/packages)
# custom module
# exception handling
# basic data visualization


import math                                     # importing math module

print("ceil value:", math.ceil(4.2))           # next greater whole number
print("floor value:", math.floor(4.8))         # lower whole number
print("square root:", math.sqrt(25))           # square root
print()

```

```python
%%writefile my_module.py
#creating my own module
def add(a, b):
    return a + b

def sub(a, b):
    return a - b

name = "own module example"
```

```python
# custom module
# here my_module is our own file
# we are importing functions from that file


import my_module                                # importing my own module file

print("module name:", my_module.name)           # taking variable from module
print("addition:", my_module.add(10, 5))        # calling add function
print("subtraction:", my_module.sub(10, 5))     # calling sub function
print()
```

```python

# exception handling
# try means check this code
# except means if error comes handle it here
# finally means this block will run anyway
try:
    a = 10
    b = 0
    print(a / b)                                # this gives division error
except ZeroDivisionError:
    print("cannot divide by zero")              # handling that error
finally:
    print("finally block always runs")          # this always runs
print()


try:
    num = int("hello")                          # string cannot become int
    print(num)
except ValueError:
    print("string cannot be converted to int")  # handling value error
finally:
    print("second example completed")           # this also always runs
print()
```

```python



# data visualization
# in class sir just showed basic plots
# matplotlib is used for graphs

import matplotlib.pyplot as plt                 # importing plotting library

# line graph ex
x = [1, 2, 3, 4, 5]                             # x values
y = [10, 20, 15, 25, 30]                        # y values

plt.plot(x, y)                                  # line graph
plt.title("line plot")                          # graph title
plt.xlabel("x values")                          # x side name
plt.ylabel("y values")                          # y side name
plt.show()                                      # showing graph


# bar chart example
students = ["A", "B", "C", "D"]                 # student names
marks = [75, 82, 68, 90]                        # marks list

plt.bar(students, marks)                        # bar graph
plt.title("student marks")                      # title
plt.xlabel("students")                          # x side
plt.ylabel("marks")                             # y side
plt.show()                                      # display graph


# pie chart example
labels = ["Python", "Java", "C++"]              # labels in pie chart
sizes = [45, 30, 25]                            # values

plt.pie(sizes, labels=labels, autopct="%1.1f%%") # pie chart with percentage
plt.title("language usage")                     # title
plt.show()                                      # display graph
```
