```python

print("1. ENCAPSULATION")
print()

class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.__balance = balance          # private variable

    def deposit(self, amount):
        self.__balance += amount

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
        else:
            print("not enough balance")

    def get_balance(self):
        return self.__balance

acc1 = BankAccount("Nikhil", 5000)
acc1.deposit(1000)
acc1.withdraw(2000)
print("account holder:", acc1.name)
print("balance:", acc1.get_balance())
# print(acc1.__balance)   # this gives error
print()
```

```python
print("2. COMPOSITION")
print()

class Engine:
    def start(self):
        print("engine started")

class Car:
    def __init__(self, brand):
        self.brand = brand
        self.engine = Engine()            # car has an engine

    def drive(self):
        print(self.brand, "car is ready")
        self.engine.start()

c1 = Car("Toyota")
c1.drive()
print()
```

```python

print("3. DYNAMIC EXTENSION")
print()

class Student:
    def __init__(self, name):
        self.name = name

s1 = Student("Anu")
s1.course = "Python"                     # adding extra attribute later
s1.marks = 92

print("name:", s1.name)
print("course:", s1.course)
print("marks:", s1.marks)
print()
```

```python


print("4. POLYMORPHISM AND DUCK TYPING")
print()

class Dog:
    def sound(self):
        print("dog barks")

class Cat:
    def sound(self):
        print("cat meows")

class Bird:
    def sound(self):
        print("bird chirps")

def make_sound(obj):
    obj.sound()                          # duck typing

make_sound(Dog())
make_sound(Cat())
make_sound(Bird())
print()
```

```python
print("5. CLASS VARIABLE VS INSTANCE VARIABLE")
print()

class Employee:
    company = "Infosys"                  # class variable

    def __init__(self, name, salary):
        self.name = name                 # instance variable
        self.salary = salary

e1 = Employee("Ravi", 40000)
e2 = Employee("Meena", 50000)

print("class variable:", Employee.company)
print("e1:", e1.name, e1.salary)
print("e2:", e2.name, e2.salary)
print()
```

```python

print("6. INSTANCE METHODS")
print()

class Circle:
    def __init__(self, radius):
        self.radius = radius

    def area(self):                      # instance method
        return 3.14 * self.radius * self.radius

c = Circle(5)
print("area:", c.area())
print()
```

```python
print("7. CLASSMETHOD")
print()

class College:
    college_name = "ABC College"

    def __init__(self, student_name):
        self.student_name = student_name

    @classmethod
    def change_college_name(cls, new_name):
        cls.college_name = new_name

    def show(self):
        print(self.student_name, "-", College.college_name)

st1 = College("Kiran")
st2 = College("Divya")

st1.show()
st2.show()

College.change_college_name("XYZ College")

st1.show()
st2.show()
print()
```

```python

print("8. STATICMETHOD")
print()

class MathTools:
    @staticmethod
    def add(a, b):
        return a + b

    @staticmethod
    def is_even(num):
        return num % 2 == 0

print("sum:", MathTools.add(10, 20))
print("is 8 even?", MathTools.is_even(8))
print()
```

```python
print("9. DECORATOR")
print()

def simple_decorator(func):
    def wrapper():
        print("function is starting")
        func()
        print("function ended")
    return wrapper

@simple_decorator
def greet():
    print("hello")

greet()
print()
```

```python

print("10. DATACLASS AND __post_init__")
print()

from dataclasses import dataclass

@dataclass
class Product:
    name: str
    price: int
    quantity: int

    def __post_init__(self):
        self.total = self.price * self.quantity    # extra work after init
        if self.price < 0:
            print("price cannot be negative")

p1 = Product("Laptop", 50000, 2)
print("name:", p1.name)
print("price:", p1.price)
print("quantity:", p1.quantity)
print("total:", p1.total)
print()
```

```python
print("11. SINGLE RESPONSIBILITY PRINCIPLE")
print()

# bad idea: one class doing too many things
class Report:
    def __init__(self, data):
        self.data = data

class ReportPrinter:
    def print_report(self, report):
        print("report data:", report.data)

class ReportSaver:
    def save_report(self, report):
        print("report saved:", report.data)

r1 = Report("monthly sales data")
rp = ReportPrinter()
rs = ReportSaver()

rp.print_report(r1)
rs.save_report(r1)
print()
```

```python

print("12. ONE MORE ENCAPSULATION EXAMPLE")
print()

class Person:
    def __init__(self, name, age):
        self._name = name                # protected style
        self.__age = age                 # private style

    def get_age(self):
        return self.__age

    def set_age(self, age):
        if age > 0:
            self.__age = age
        else:
            print("invalid age")

p = Person("Sai", 22)
print("name:", p._name)
print("age:", p.get_age())
p.set_age(25)
print("new age:", p.get_age())
print()


print("assignment completed")
```
